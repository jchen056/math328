Interpolation: you have measurements that are assumed to be very reliable; you want an estimation function that goes through the data points you have

n-th degree polynomial $P_n(x)=c_0+c_1x+...+c_nx^n$ has n+1 parameters: $c_0$, $c_1$,...,$c_n$

In other words, given any function and any n + 1 points x in the domain of the function, we can find an nth degree polynomial which agrees with the function at each of those n + 1 values of x.

1. Brute force approach(direct method)
2. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.interpolate as interpolate
import scipy
from numpy.polynomial.polynomial import Polynomial
from IPython.display import Image
```

```python
#direct approach: given data points(3,32),(5,74), and (7,132)
#find a quadratic function that pass through all the points
A= np.array([[1, 3,9], [1, 5,25],[1,7,49]])
b=np.array([32,74,132])
x = np.linalg.solve(A, b)
print("c0+c1x+c2x^2 with c0, c1, and c2=")
print(x)

P2=Polynomial(x)
plt.plot(np.arange(2,8,0.1),P2(np.arange(2,8,0.1)),'b', np.array([3, 5, 7]),np.array([32, 74, 132]),'ro')
plt.title('Direct method to Polynomial Interpolation')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Lagrance polynomial
Lagrange polynomial interpolation finds a single polynomial that goes through all the data points. This polynomial is referred to as a Lagrange polynomial.


```python
x=np.array([3, 5, 7])
y=np.array([32, 74, 132])
lagrange_interpolation = scipy.interpolate.lagrange([3, 5, 7], [32, 74, 132])#Return a Lagrange interpolating polynomial
plt.plot(np.arange(2,8,0.1),lagrange_interpolation(np.arange(2,8,0.1)), 'b', x, y, 'ro',)

plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def _poly_newton_coefficient(x, y):    
    m = len(x)
    x = np.copy(x)
    a = np.copy(y)
    for k in range(1, m):
        a[k:m] = (a[k:m] - a[k - 1])/(x[k:m] - x[k - 1])
    return a

def newton_polynomial(x_data, y_data, x):
    a = _poly_newton_coefficient(x_data, y_data)
    n = len(x_data) - 1  # Degree of polynomial
    p = a[n]
    for k in range(1, n + 1):
        p = a[n - k] + (x - x_data[n - k])*p
    return p
```

```python
plt.plot(x, y,'bo')
plt.plot(np.arange(2,8,0.1),newton_polynomial(x, y, np.arange(2,8,0.1)))
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title("Newton interpolation")
```
