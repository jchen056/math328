```python
import pandas as pd
import numpy as np
```

Model the spread of disease:

1. Susceptible: those who have never had the illness and can catch it;
2. Infected: those who currently have the illness and are contagious;
3. Recovered: those who have already had the illness and are immune

a: transmission coefficient; depends on the general health for all populations and the level of social interaction
b: depends on the illness involved(I->R)

1. S'=-aSI
2. I'=aSI-bI
3. R'=bI

```python
#intial condition
S=45400
I=2100
R=2500
a=0.00001
b=1/14
def S_change(a,S,I):
    return -a*S*I

def I_change(a,b,S,I):
    return a*S*I-b*I

def R_change(b,I):
    return b*I
```

```python
df_future=pd.DataFrame(np.zeros((4,6)),columns=['S','I','R',"S'","I'","R'"])
df_future.iloc[0]=S,I,R,S_change(a,S,I),I_change(a,b,S,I),R_change(b,I)
df_future
```

```python
for i in range(1,4):
    S_updated=df_future.iloc[i-1]['S']+df_future.iloc[i-1]["S'"]*1
    I_updated=df_future.iloc[i-1]['I']+df_future.iloc[i-1]["I'"]*1
    R_updated=df_future.iloc[i-1]['R']+df_future.iloc[i-1]["R'"]*1
    
    Sp_updated=S_change(a,S_updated,I_updated)
    Ip_updated=I_change(a,b,S_updated,I_updated)
    Rp_updated=R_change(b,I_updated)
    df_future.iloc[i]=S_updated,I_updated,R_updated,Sp_updated,Ip_updated,Rp_updated

df_future
```
