Fourier Transform: express some complicated functions as the infinite sum of sine and cosine waves.

We can model a single wave as a field with a function 𝐹(𝑥,𝑡), where 𝑥 is the location of a point in space, while 𝑡 is the time.
1. Amplitude: used to describe the difference between the maximum values to the baseline value
2. period: time it takes to finish the complete cycle; measured from the two adjacent peaks
3. Wavelength: the distance between two successive crests or troughs of a wave
4. Frequency: the number of waves that pass a fixed place in a given amount of time; freqeuncy=1/period

𝑦(𝑡)=𝐴𝑠𝑖𝑛(𝜔𝑡+𝜙)
where 𝐴 is the amplitude of the wave, 𝜔 is the angular frequency, which specifies how many cycles occur in a second, in radians per second. 𝜙 is the phase of the signal. If 𝑇 is the period of the wave, and 𝑓 is the frequency of the wave, then 𝜔 has the following relationship to them:

𝜔=2𝜋/𝑇=2𝜋𝑓

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
x = np.linspace(0, 20, 201)
y = np.sin(x)

plt.figure(figsize = (8, 6))
plt.plot(x, y, 'b')
plt.ylabel('Amplitude')
plt.xlabel('Location (x)')
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))

times = np.arange(5)

n = len(times)

for t in times:
    plt.subplot(n, 1, t+1)
    y = np.sin(x + t)#sine wave can change both in time and space
    plt.plot(x, y, 'b')
    plt.plot(x[25], y [25], 'ro')
    plt.ylim(-1.1, 1.1)
    plt.ylabel('y')
    plt.title(f't = {t}')

plt.xlabel('location (x)')
plt.tight_layout()
plt.show()
```

generate two sine waves with time between 0 and 1 seconds and frequency is 5 Hz and 10 Hz, all sampled at 100 Hz. Plot the two waves and see the difference. Count how many cycles in the 1 second.

Usually when we record a wave, we need to specify how often we sample the wave in time, this is called sampling. And this rate is called sampling rate, with the unit Hz. For example, if we sample a wave at 2 Hz, it means that every second we sample two data points. 

```python
# sampling rate
sr = 100.0
# sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)

# frequency of the signal
freq = 5   
y = np.sin(2*np.pi*freq*t)

plt.figure(figsize = (8, 8))
plt.subplot(211)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

freq = 10   
y = np.sin(2*np.pi*freq*t)

plt.subplot(212)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

plt.xlabel('Time (s)')
plt.show()
```

Generate two sine waves with time between 0 and 1 seconds. Both waves have frequency 5 Hz and sampled at 100 Hz, but the phase at 0 and 10, respectively. Also the amplitude of the two waves are 5 and 10. Plot the two waves and see the difference.

```python
# frequency of the signal
freq = 5   
y = 5*np.sin(2*np.pi*freq*t)

plt.figure(figsize = (8, 8))
plt.subplot(211)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

y = 10*np.sin(2*np.pi*freq*t + 10)

plt.subplot(212)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

plt.xlabel('Time (s)')
plt.show()
```
